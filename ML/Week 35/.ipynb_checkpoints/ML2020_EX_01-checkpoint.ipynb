## 2 Eigendecomposition

Eigenvalues and eigenvectors of matrix $A$ can be extracted using the Numpy function:
```python
Î», V = numpy.linalg.eig(A)
```
where $\lambda$ is a vector of eigenvalues, and $V$ is a matrix of eigenvectors. 

### 2.1 For matrix A, calculate  $\lambda$ and $V$.
\begin{equation*}
A = 
\begin{pmatrix}
1 & 5   \\
3 & 8 
\end{pmatrix}
\end{equation*}


#### Check if  for each $\lambda_i$ and its corresponding verctor $ \vec{v_i}$ the equation $ A\vec{v_i} = \lambda_i \vec{v_i}$ stands.

Note that each column of $ V $ is an eigenvector of $ A $. 




### 2.2 Write a reverse process that calculates the original matrix from the given eigenvalues and eigenvectors.


\begin{equation*}
A = V \Lambda V^{-1}
\end{equation*}

where $\Lambda$ is a diagonal matrix that its diagonal elements are the eigenvalues.


````python
import numpy as np

A = np.array([[1,5],[3,8]])


lam, V = np.linalg.eig(A)




# print(lam)

# print("------------")

# print(V)

# print("----")
# print(lam[0])


# print("Product is")
print(lam[0]*V[:,0])
# print("Product are")
print(np.dot(A,V[:,0]))

print("It holds")

print(lam[1]*V[:,1])
# print("Product are")
print(np.dot(A,V[:,1]))

print("It holds")


````

````python
D = np.diag(lam)

inv = np.linalg.inv(V)

ans = np.dot(V, D)

ans = np.dot(ans, inv)

print(ans)

print("Which is the original matrix")
````
